```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd
from sklearn.impute import SimpleImputer

# Charger dataset
df = kagglehub.dataset_load( # Using dataset_load as recommended
    KaggleDatasetAdapter.PANDAS,
    "tarktunataalt/breast-tissue-impedance-measurements",
    "data.csv" # Corrected file name to data.csv
)

print("Shape:", df.shape)
print("Classes:", df["Class"].unique())

# Nettoyage simple
imputer = SimpleImputer(strategy="mean")
df_clean = df.copy()
df_clean.iloc[:, 1:] = imputer.fit_transform(df.iloc[:, 1:])  # sauf colonne Class

print("\nStatistiques :")
print(df_clean.describe())

print("\n5 premières lignes propres :")
print(df_clean.head())
```

## Complete data cleaning and display first 5 rows

### Subtask:
Complete the cleaning steps and display the first 5 rows of the cleaned feature set (X_clean).


**Reasoning**:
The subtask requires separating features and target variables, encoding the target variable, and then displaying the first 5 rows of the cleaned features. This code block will perform these actions sequentially.



```python
from sklearn.preprocessing import LabelEncoder

# Separate features (X) and target variable (y)
X_clean = df_clean.drop('Class', axis=1)
y = df_clean['Class']

# Encode categorical 'Class' labels into numerical labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print("First 5 rows of X_clean:")
print(X_clean.head())
```

## Split data into training and testing sets

### Subtask:
Split the cleaned data into training and testing sets (X_train, X_test, y_train, y_test) using a 80/20 split and a random state of 42.


**Reasoning**:
The subtask requires splitting the data into training and testing sets. I will import the necessary function, perform the split with the specified parameters, and then verify the shapes of the resulting datasets.



```python
from sklearn.model_selection import train_test_split

# Split the cleaned data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_encoded, test_size=0.2, random_state=42)

# Print the shapes of the new datasets to verify the split
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

## Train a RandomForestClassifier model

### Subtask:
Initialize and train a RandomForestClassifier model with a random state of 42 on the training data.


**Reasoning**:
The subtask requires training a RandomForestClassifier model. I will import the necessary class, instantiate it with the specified random state, and then fit it to the training data (X_train and y_train).



```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the RandomForestClassifier model with random_state=42
model = RandomForestClassifier(random_state=42)

# Fit the model to the training data
model.fit(X_train, y_train)

print("RandomForestClassifier model trained successfully.")
```

## Evaluate model performance and visualize Confusion Matrix

### Subtask:
Make predictions on the test set and evaluate the model's performance using accuracy score, classification report, and then visualize the confusion matrix to show the performance across different classes.


**Reasoning**:
I need to make predictions on the test set, calculate accuracy, generate a classification report, and visualize the confusion matrix as per the instructions. This involves importing evaluation metrics and plotting functions from `sklearn.metrics` and utilizing the `label_encoder` from a previous step.



```python
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 1. Use the trained model to make predictions on the test set
y_pred = model.predict(X_test)

# 3. Calculate the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: {accuracy:.2f}")

# 4. Generate and print the classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# 5. Create a list of class names
class_names = label_encoder.classes_

# 6. Plot the confusion matrix
fig, ax = plt.subplots(figsize=(8, 6))
cm_display = ConfusionMatrixDisplay.from_predictions(
    y_test,
    y_pred,
    display_labels=class_names,
    cmap=plt.cm.Blues,
    ax=ax
)

# 7. Add a title to the confusion matrix plot
ax.set_title('Confusion Matrix')

# 8. Display the plot
plt.show()
```

## Visualize Feature Importance

### Subtask:
Extract and visualize the feature importances from the trained RandomForestClassifier to understand which features contribute most to the model's predictions.


**Reasoning**:
I need to extract feature importances from the trained RandomForestClassifier model, associate them with feature names, sort them, and then visualize them using a bar plot. This will help understand which features are most influential in the model's predictions.



```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Extract feature importances from the trained model
feature_importances = model.feature_importances_

# 2. Create a Pandas Series to store feature names and their corresponding importances
feature_names = X_clean.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})

# 3. Sort the features by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# 4. Create a bar plot to visualize the feature importances
plt.figure(figsize=(10, 6))
plt.bar(feature_importance_df['Feature'], feature_importance_df['Importance'], color='skyblue')

# 5. Add labels and title
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importances from RandomForestClassifier')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# 6. Display the plot
plt.show()
```
